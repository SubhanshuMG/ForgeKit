```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data here
# df = pd.read_csv('../data/raw/your_data.csv')

# Synthetic example
np.random.seed(42)
df = pd.DataFrame({'feature_1': np.random.randn(100), 'feature_2': np.random.randn(100), 'target': np.random.randint(0, 2, 100)})
print(df.shape)
df.head()
```

```python
df.describe()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
df[['feature_1', 'feature_2']].hist(ax=axes[0])
df['target'].value_counts().plot(kind='bar', ax=axes[1])
plt.tight_layout()
plt.show()
```
